```python
import requests
import pickle
import os
```

```python
def get_user_submissions(handle):
    # The API endpoint for user status
    url = f"https://codeforces.com/api/user.status?handle={handle}"
    
    try:
        response = requests.get(url)
        data = response.json() # Converts the raw text into a Python dictionary
        
        if data["status"] == "OK":
            # TASK: Look at the 'data' dictionary. 
            # Codeforces returns the list of submissions under the key "result".
            submissions = data["result"] 
            return submissions
        else:
            print(f"Error: {data['comment']}")
            return []
            
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Test it with a known handle
user_handle = "E_Abishek" # Feel free to change this!
submissions = get_user_submissions(user_handle)
print(f"Fetched {len(submissions)} submissions for {user_handle}.")
```

```python
def get_solved_problems(submissions):
    solved_problems = set() # Using a set for fast lookups
    
    for sub in submissions:
        # Check if the submission was successful
        if sub.get('verdict') == 'OK':
            problem = sub['problem']
            # Create a unique ID like "2211H"
            problem_id = f"{problem['contestId']}{problem['index']}"
            solved_problems.add(problem_id)
            
    return solved_problems

# usage:
solved_set = get_solved_problems(submissions)
print(f"User has solved {len(solved_set)} unique problems.")
```

```python
def get_user_skill_map(submissions):
    skill_map = {}
    
    for sub in submissions:
        if sub.get('verdict') == 'OK':
            tags = sub['problem'].get('tags', [])
            for tag in tags:
                # Increment the count for each tag found
                skill_map[tag] = skill_map.get(tag, 0) + 1
                
    return skill_map

# Example output might look like: {'dp': 15, 'math': 10, 'graphs': 2}
user_skills = get_user_skill_map(submissions)
print(user_skills)
```

```python
def calculate_recommendation_score(user_skills, problem_tags):
    score = 0
    for tag in problem_tags:
        # If the user has experience in this tag, add their 'points' to the score
        score += user_skills.get(tag, 0)
    return score

# Example usage:
unsolved_problem_tags = ['math', 'geometry']
# If user_skills has {'math': 981, 'geometry': 97}, the score would be 1078.
score = calculate_recommendation_score(user_skills, unsolved_problem_tags)
print(f"Recommendation Score: {score}")
```

```python
# 1. Sample 'Available' Problems (Usually you'd fetch this from /api/problemset.problems)
available_problems = [
    {"name": "Dynamic Math Challenge", "id": "101A", "tags": ["dp", "math"], "rating": 1400},
    {"name": "The Great Graph", "id": "102B", "tags": ["graphs", "dfs and similar"], "rating": 1600},
    {"name": "Bitwise Fun", "id": "103C", "tags": ["bitmasks", "constructive algorithms"], "rating": 1200},
    {"name": "String Master", "id": "104D", "tags": ["strings", "hashing"], "rating": 1500}
]

def get_recommendation(user_skills, problem_list):
    best_score = -1
    best_problem = None
    
    for prob in problem_list:
        # Calculate score based on Phase 4 logic
        score = sum(user_skills.get(tag, 0) for tag in prob['tags'])
        
        print(f"Checking {prob['name']}... Score: {score}")
        
        if score > best_score:
            best_score = score
            best_problem = prob
            
    return best_problem, best_score

# Run the recommendation
top_prob, top_score = get_recommendation(user_skills, available_problems)

print("-" * 30)
print(f"🚀 TOP RECOMMENDATION: {top_prob['name']} ({top_prob['id']})")
print(f"Why? It matches your skills with a score of {top_score}!")
```

```python
import requests
import pickle
import os

def get_categorized_recommendations(user_handle, target_tags):
    # 1. Setup User Data & Current Rating
    # Fetch submissions for solved set and skill calculation
    sub_url = f"https://codeforces.com/api/user.status?handle={user_handle}"
    submissions = requests.get(sub_url).json().get('result', [])
    
    # Fetch User Info to get current rating
    user_info_url = f"https://codeforces.com/api/user.info?handles={user_handle}"
    user_data = requests.get(user_info_url).json().get('result', [{}])[0]
    current_rating = user_data.get('rating', 1200)  # Default to 1200 if unrated
    
    print(f"Targeting problems for rating: {current_rating} ± 200")

    solved_set = {f"{s['problem']['contestId']}{s['problem']['index']}" 
                  for s in submissions if s.get('verdict') == 'OK'}

    user_skills = {}
    for s in submissions:
        if s.get('verdict') == 'OK':
            for tag in s['problem'].get('tags', []):
                user_skills[tag] = user_skills.get(tag, 0) + 1

    # 2. Fetch Global Problem Set — WITH PICKLE CACHE
    CACHE_FILE = "cf_problems_cache.pkl"
    if os.path.exists(CACHE_FILE):
        with open(CACHE_FILE, "rb") as f:
            all_probs = pickle.load(f)
    else:
        all_probs = requests.get("https://codeforces.com/api/problemset.problems").json()['result']['problems']
        with open(CACHE_FILE, "wb") as f:
            pickle.dump(all_probs, f)

    # 3. Categorize & Filter by Rating
    categorized_recs = {tag: [] for tag in target_tags}
    
    # Define rating bounds
    lower_bound = current_rating - 200
    upper_bound = current_rating + 200

    for prob in all_probs:
        prob_id = f"{prob['contestId']}{prob.get('index', '')}"
        prob_rating = prob.get('rating')

        # FILTER: Skip if solved OR if rating is outside ±200 range
        if prob_id in solved_set:
            continue
        if prob_rating is None or not (lower_bound <= prob_rating <= upper_bound):
            continue

        p_tags = prob.get('tags', [])
        score = sum(user_skills.get(t, 0) for t in p_tags)
        
        for tag in target_tags:
            if tag in p_tags:
                categorized_recs[tag].append({
                    'name': prob['name'],
                    'rating': prob_rating,
                    'url': f"https://codeforces.com/contest/{prob['contestId']}/problem/{prob['index']}",
                    'score': score
                })

    # 4. Sort and pick top 2
    final_output = {}
    for tag in target_tags:
        # Sort by user skill score; you could also sort by 'rating' if preferred
        sorted_list = sorted(categorized_recs[tag], key=lambda x: x['score'], reverse=True)
        final_output[tag] = sorted_list[:2]

    return final_output

# EXECUTION
my_topics = ["math", "dp", "recursion", "graphs", "data structures", 
             "binary search", "two pointers", "greedy", "strings", "bitmasks"]

# Example usage:
# recommendations = get_categorized_recommendations("your_handle_here", my_topics)
```

```python
results = get_categorized_recommendations("E_Abishek", my_topics)
for topic, probs in results.items():
    print(f"--- Topic: {topic.upper()} ---")
    for p in probs:
        print(f" - {p['name']} (Score: {p['score']})")
        print(f"   {p['url']}")
    print()
```
